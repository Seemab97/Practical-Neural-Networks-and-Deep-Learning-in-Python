# Data Imputation

If we keep removing everything, data size will decrease. Here comes data imputation. 

Age is a quantitative variable. We can replace missing age values with the average age values.

### Replace NANs

```python
d3=df.copy(deep=True)
d3.head(6) #from output above, we know we need to explore till index 5 to see NaN in age
```

```python
#fillna() instead of dropna()

d3_imp_mean=d3['Age'].fillna(d3['Age'].mean()) #median can also be used
d3_imp_mean.head(6) #in the output above, age is NaN at index 5, it should be replaced with mean age
```

```python
d3_imp_mean.loc[5] #original dataframe has age NaN at index 5
```
